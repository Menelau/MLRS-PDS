```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('dataframe_friedman.csv')
#removing majority voting performance
df.drop(list(df.filter(regex='MV')), axis=1, inplace=True)
```

# FLT

```python
max_count = df.filter(regex='FLT', axis=1).idxmax(axis=1).value_counts()
print(max_count)
print(max_count*100/288)
max_count.plot(kind='bar')
plt.title('Highest Result Count per Method')
plt.xlabel('Method')
plt.ylabel('Count of Highest Results')
plt.xticks(rotation=90)
plt.tight_layout()
plt.savefig('FLT.pdf',dpi=450)
plt.show()
```
